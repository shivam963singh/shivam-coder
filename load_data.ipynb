below code help us to merge both data set's that help us to do our work

```python
merged_tt = pd.merge(emp,dep,on="Department_id",how= "outer")
print(merged_tt)
```

```python
merged_tt.head(10)
merged_tt.to_csv(r"C:\Users\Homes\OneDrive\Desktop\data projects\Joined_Table.csv")
```

in this code we are getting all employe that are working in ranchi there name age and salary

```python
pt = merged_tt[merged_tt["Location"] == "Ranchi"]
print(pt[["first_name","Last_name","age","salary","Department_name"]])
pt.to_csv(r"C:\Users\Homes\OneDrive\Desktop\data projects\Ranchi_employee.csv")
```

top 10 highest salary   employee details 

```python
highest_salary = merged_tt.sort_values(by="salary", ascending=False).head(10)
print(highest_salary[["first_name", "Last_name", "age", "salary", "Department_name"]])
highest_salary.to_csv(r"C:\Users\Homes\OneDrive\Desktop\data projects\highest_salary.csv")
```

list of employee who are freshers 

```python
freshers = merged_tt.query('age <= 25' ).sort_values(by='first_name')
print(freshers[["first_name","Last_name","age","salary","Department_name"]])
freshers.to_csv(r"C:\Users\Homes\OneDrive\Desktop\data projects\freshers.csv")
```
